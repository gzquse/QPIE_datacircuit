```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import jn  # Bessel function of the first kind

# Define x values
x = np.linspace(0, 20, 500)

# Compute J0(x) and J1(x)
J0 = jn(0, x)
J1 = jn(1, x)

# Plot the functions
plt.plot(x, J0, label='J0(x)')
plt.plot(x, J1, label='J1(x)')
plt.title("Bessel Functions of the First Kind")
plt.xlabel("x")
plt.ylabel("J_n(x)")
plt.legend()
plt.grid()
plt.show()
```
